## Coding Exercise 10: Plotting from Pandas Dataframes

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd
```

```python
df=pd.read_excel("verlegenhuken.xlsx",0)
df.head()
```

```python
p=figure(plot_width=1000,plot_height=800) 
p.title.text="Temperature & Air Pressure"
p.title.text_font="times"
p.title.text_font_size="18pt"
p.title.align="center"

X=df["Temperature"]/10
Y=df["Pressure"]/10

p.xaxis.axis_label="Temperature [Â°C]"
p.yaxis.axis_label="Air Pressure [hPa]"

p.circle(X,Y, color="red", alpha=0.5) #define how to show the datapoint

output_file("Scatter_plotting.html")
show(p)
```
