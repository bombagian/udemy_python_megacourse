## Notes and code from the Udemy lecture on Data Visualisation

3 interfaces for visualisation with Bokeh:

-  bokeh.models: high flexibility/customisation - for developers 
-  bokeh.plotting: mid customisation (suggested)
-  bokeh.charts: limited customisation. Quicker but less flexible

```python
#firt example from the easiest interface, bokeh.charts
from bokeh.charts import Scatter, output_file, show
import pandas as pd

#create dataframe manually
df=pd.DataFrame(columns=["X","Y"])
df["X"]=[1,2,3,4,5]
df["Y"]=[5,6,4,5,3]


p = Scatter(df,x="X",y="Y",title="Temperature Observations",xlabel="Day of observations",ylabel="Temperature")
output_file("Scatter charts.html")
show(p)



```

```python
#replicate the above example with the plotting interface

from bokeh.plotting import figure, output_file, show


#p=figure(plot_width=500,plot_height=400, title="title") 
# instead of put attributes to the figure method, it is reccomended to use the p.method approach as below: 

p=figure(plot_width=500,plot_height=400) 

p.title.text="TITLE"
p.title.text_color="Red"
p.title.text_font="times"
p.title.text_font_style="italic"
p.title.text_font_size="18pt"
p.title.align="center"

#Use HELP to explore the attributes and methods available
#help(p) 

#add more customisation for the axis
p.xaxis.axis_label="pippo"
p.yaxis.axis_label="pluto"
p.yaxis.minor_tick_line_color="red"

#create the empty figure"
#p.circle([1,2,3,4,5],[5,6,4,5,3], size=12, color="red", alpha=0.5) #define how to show the datapoint
#p.triangle([1,2,3,4,5],[5,6,4,5,3], size=12, color="blue", alpha=0.5) #define how to show the datapoint


#you can specify the size for each "point". Size is a list that matches the order of the coordinates (e.g. point(1,10) will have size 10, point (2,6) size 12, etc)
#p.circle([1,2,3,4,5],[5,6,4,5,3], size=[10,12,5,20,5], color="red", alpha=0.5) #define how to show the datapoint

#double the size using list comprehension
p.circle([1,2,3,4,5],[5,6,4,5,3], size=[i*2 for i in [10,12,5,20,5]], color="red", alpha=0.5) #define how to show the datapoint

output_file("Scatter_plotting.html")
show(p)
 
    
    
```
